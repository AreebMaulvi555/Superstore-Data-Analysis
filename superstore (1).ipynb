```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (Replace with your actual dataset)

# Step 1: Create a pivot table to track customer presence in each year
customer_year_matrix = df.pivot_table(index="Customer Name", columns="order_year", aggfunc="size", fill_value=0)

# Step 2: Convert to binary (1 if customer ordered in that year, else 0)
customer_year_matrix = (customer_year_matrix > 0).astype(int)

# Step 3: Compute total customers per year
total_customers = customer_year_matrix.sum()

# Step 4: Compute retained customers (customers appearing in previous and current year)
retention = customer_year_matrix.cumsum(axis=1)
retained_customers = (retention > 1).sum()

# Step 5: Plot retention over time
plt.figure(figsize=(8, 5))
plt.plot(total_customers.index, total_customers.values, marker="o", label="Total Customers", linestyle="--")
plt.plot(retained_customers.index, retained_customers.values, marker="s", label="Retained Customers", linestyle="-")

plt.xlabel("Year")
plt.ylabel("Number of Customers")
plt.title("Customer Retention Over Time (2014-2017)")
plt.legend()
plt.grid(True)
plt.show()
```
